#List

```python
# Membuat list
angka = [10, 20, 30, 40, 50]  # List berisi angka

# Menambahkan item
angka.append(60)       # Menambahkan 60 di akhir list
angka.insert(2, 25)    # Menambahkan 25 di index ke-2

# Menghapus item
angka.remove(20)       # Menghapus nilai 20
angka.pop(0)           # Menghapus item index 0 (10)

# Mengakses item / slicing
print("List akhir:", angka)
print("Item index 1:", angka[1])
print("Slice index 1-3:", angka[1:4])
```

#Manipulasi List & Basic Calculation

```python
# List awal
angka = [25, 30, 40, 50, 60]

# Menghitung total dengan loop
total = 0
for x in angka:
    total += x   # total = total + x
print("Total angka:", total)

# Menghitung rata-rata
rata_rata = total / len(angka)  # len() = jumlah item di list
print("Rata-rata:", rata_rata)

# Nilai maksimum & minimum
maks = max(angka)  # fungsi max() cari nilai terbesar
minim = min(angka) # fungsi min() cari nilai terkecil
print("Nilai maksimum:", maks)
print("Nilai minimum:", minim)

# Membuat list baru berisi kuadrat tiap angka (list comprehension)
kuadrat = [x**2 for x in angka]  # x**2 = x pangkat 2
print("Kuadrat tiap angka:", kuadrat)
```

Mini Project

```python
# Dataset mini: nama siswa dan nilai
siswa = ["Dapala", "Rina", "Budi", "Sari"]
nilai = [80, 65, 90, 55]

# Fungsi untuk hitung total
def hitung_total(list_nilai):
    total = 0
    for x in list_nilai:
        total += x
    return total

# Fungsi untuk rata-rata
def hitung_rata(list_nilai):
    return hitung_total(list_nilai) / len(list_nilai)

# Fungsi untuk kategori
def kategori(nilai_siswa):
    if nilai_siswa >= 70:
        return "Lulus"
    else:
        return "Tidak Lulus"

# Hitung total & rata-rata
total_nilai = hitung_total(nilai)
rata_nilai = hitung_rata(nilai)
maks_nilai = max(nilai)
min_nilai = min(nilai)

print("Total nilai:", total_nilai)
print("Rata-rata nilai:", rata_nilai)
print("Nilai maksimum:", maks_nilai)
print("Nilai minimum:", min_nilai)

# Tentukan kategori tiap siswa
for i in range(len(siswa)):
    print(siswa[i], "-", nilai[i], "-", kategori(nilai[i]))

# Buat list baru: nilai kuadrat
nilai_kuadrat = [x**2 for x in nilai]
print("Nilai kuadrat:", nilai_kuadrat)
```
